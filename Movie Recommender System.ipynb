```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('tmdb_5000_movies.csv')
df2=pd.read_csv('tmdb_5000_credits.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
data=df1.merge(df2,on='title')
data.head()
```

```python
data=data[['id','overview','genres','keywords','cast','crew','title']]
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data.shape
```

```python
data.duplicated().sum()
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres']=data['genres'].apply(convert)
data['keywords']=data['keywords'].apply(convert)
```

```python
def conv(obj):
    c=0
    l=[]
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c=c+1
        else:
            break
    return l
```

```python
data['cast']=data['cast'].apply(conv)
```

```python
def conver(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew']=data['crew'].apply(conver)
```

```python
data['overview']=data['overview'].apply(lambda x:x.split())
```

```python
data['tags']=data['overview']+data['genres']+data['keywords']+data['cast']+data['crew']
data.drop(columns=['overview','genres','keywords','cast','crew'],axis=1,inplace=True)
data.head()
```

```python
data['tags']=data['tags'].apply(lambda x:[i.lower() for i in x])
data['tags']=data['tags'].apply(lambda x:[i.replace(" ","") for i in x])
data.head()
```

```python
data['tags']=data['tags'].apply(lambda x:" ".join(x))
data.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from autocorrect import Speller
spell = Speller(fast = True)
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
import contractions
import regex as re


def cleanup(sentence):
    sentence = re.sub("[^a-zA-Z]", " ", str(sentence))
    sentence = spell(sentence)
    sentence = contractions.fix(sentence)
    sentence = sentence.split()
    sentence = [word for word in sentence if word not in stop_words]
    sentence = [ps.stem(token) for token in sentence]

    return (" ".join(sentence))
```

```python
data['tags']=data['tags'].apply(cleanup)
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000)
```

```python
vectors=cv.fit_transform(data['tags']).toarray()
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    index=data[data['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(data['title'][i[0]])
```
